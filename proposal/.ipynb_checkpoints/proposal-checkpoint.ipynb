```r
library(tidyverse)
library(repr)
library(tidymodels)
library(cowplot)
options(repr.matrix.max.rows = 6)
```

```r
sessions <- read_csv("project/data/sessions.csv")

sessions
```

```r
start_plot <- ggplot(sessions, aes(x = original_start_time)) +
    geom_bar()
start_plot
```

```r
end_plot <- ggplot(sessions, aes(x = original_end_time)) +
    geom_bar()
end_plot
```
